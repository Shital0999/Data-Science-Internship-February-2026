```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    # Rule 1: Minimum recharge amount
    if amount < 50:
        print("Recharge amount must be at least ₹50")
        return False

    # Rule 2: Check valid plans
    if amount not in valid_plans:
        print("Invalid plan. Valid plans are:", valid_plans)
        return False

    return True


# Retry using while loop
while True:
    amount = int(input("Enter recharge amount: ₹"))

    if validate_recharge(amount):
        print("Recharge successful")
        break
    else:
        print("Please try again\n")
```

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert (Stock = {quantity})")
        else:
            print(f"{product}: Stock OK (Stock = {quantity})")


# Example inventory dictionary
inventory = {
    "Rice": 25,
    "Sugar": 10,
    "Oil": 18,
    "Salt": 8,
    "Flour": 30
}

# Call function
check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System

def process_result(marks):
    total = 0

    # Calculate total using loop
    for m in marks:
        total += m

    # Calculate average
    average = total / len(marks)

    # Determine result
    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"

    return average, result


# Example marks list
marks = [60, 45, 70, 55, 40]

avg, res = process_result(marks)

print("Average Marks:", avg)
print("Result:", res)
```

```python
# Problem 4: Cab Fare Estimator with Retry Option

def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    # Basic fare
    fare = base_fare + (distance * per_km_rate)

    # Peak hour extra
    if peak.lower() == "yes":
        fare = fare + (fare * 0.25)

    return fare


# Retry using while loop
while True:
    distance = float(input("Enter distance (km): "))
    peak = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(distance, peak)

    print(f"Estimated Fare: ₹{total_fare:.2f}")

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        print("Thank you!")
        break
```

```python
# Problem 5: Employee Attendance Eligibility Checker

def check_eligibility(attendance):
    present_count = 0

    # Count present days using loop
    for day in attendance:
        if day.upper() == "P":
            present_count += 1

    # Calculate percentage
    percentage = (present_count / len(attendance)) * 100

    # Determine eligibility
    if percentage >= 75:
        result = "Eligible"
    else:
        result = "Not Eligible"

    return percentage, result


# Example attendance list
attendance_list = ["P", "P", "A", "P", "P", "A", "P", "P"]

percent, status = check_eligibility(attendance_list)

print(f"Attendance Percentage: {percent:.2f}%")
print("Status:", status)
```

```python
# Problem 6: Password Strength Checker

def check_password(password):
    has_digit = False
    has_special = False

    # Check each character using loop
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    # Conditions
    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


# Example
pwd = input("Enter password: ")

result = check_password(pwd)

print("Result:", result)
```
