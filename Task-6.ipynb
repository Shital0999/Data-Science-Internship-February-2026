```python
# Problem Statement 1:Smart Parking Lot Management System

def smart_parking(capacity, logs):
    current = 0
    peak = 0

    for log in logs:
        if log.upper() == "IN":
            current += 1

        elif log.upper() == "OUT":
            if current > 0:   # prevent negative
                current -= 1

        # track peak usage
        if current > peak:
            peak = current

    # check capacity alert
    if current > capacity:
        status = "Alert: Parking Exceeded Capacity"
    elif current == capacity:
        status = "Parking Full"
    else:
        status = "Available"

    # output
    print("Currently Parked Vehicles:", current)
    print("Peak Parking Usage:", peak)
    print("Parking Status:", status)



# main program

capacity = int(input("Enter Parking Capacity: "))

logs_input = input("Enter Vehicle Logs separated by space (IN/OUT): ")
logs = logs_input.split()

smart_parking(capacity, logs)
```

```python
# Problem Statement 2:Online Food Delivery Time Estimator

def estimate_delivery_time(distance, traffic, weather):
    # base speed (km per hour)
    speed = 30

    # base time in minutes
    base_time = (distance / speed) * 60

    delay = 0

    # traffic delay
    traffic = traffic.lower()
    if traffic == "low":
        delay += 5
    elif traffic == "medium":
        delay += 10
    elif traffic == "high":
        delay += 20

    # weather delay
    weather = weather.lower()
    if weather == "rainy":
        delay += 15
    elif weather == "storm":
        delay += 25

    final_time = base_time + delay

    print("Estimated Delivery Time:", round(final_time), "minutes")


# main program

distance = float(input("Enter Distance (km): "))
traffic = input("Enter Traffic Level (Low / Medium / High): ")
weather = input("Enter Weather (Clear / Rainy / Storm): ")

estimate_delivery_time(distance, traffic, weather)
```

```python
# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer


def analyze_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)

    occupancy = (booked_count / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 70:
        status = "Almost Full"
    else:
        status = "Seats Available"

    if occupancy >= 85:
        suggestion = "Consider opening additional shows"
    else:
        suggestion = "No extra show needed"

    print("Occupancy:", round(occupancy), "%")
    print("Show Status:", status)
    print("Suggestion:", suggestion)


# MAIN
total_seats = int(input("Enter Total Seats: "))

booked_input = input("Enter booked seats (example: 1 1 1 1 ...): ")
booked_seats = list(map(int, booked_input.split()))

analyze_occupancy(total_seats, booked_seats)
```

```python
# Problem Statement 4:Cloud Server Load Classification System

def classify_server_load(cpu_readings):
    total = 0

    # loop to calculate sum
    for value in cpu_readings:
        total += value

    avg = total / len(cpu_readings)

    # classification
    if avg < 50:
        status = "Normal"
    elif avg <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", round(avg), "%")
    print("Server Status:", status)

readings_input = input("Enter CPU readings separated by space: ")
cpu_readings = list(map(float, readings_input.split()))

classify_server_load(cpu_readings)
```

```python
# Problem Statement 5:Smart Classroom Resource Usage Monitor

def monitor_resources(resource_usage):
    threshold = 8
    overused = []

    # check usage
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused.append(resource)

    # output
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

resource_usage = {}

n = int(input("Enter number of resources: "))

for i in range(n):
    name = input("Enter resource name: ")
    hours = float(input("Enter usage hours: "))
    resource_usage[name] = hours

monitor_resources(resource_usage)
```

```python
# Problem Statement 6: Online Event Registration Capacity Controller

def manage_registrations(capacity, registrations):
    confirmed = 0
    waitlisted = 0

    # simulate registrations using loop
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    # status
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"

    # output
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)
    

capacity = int(input("Enter Event Capacity: "))
registrations = int(input("Enter Total Registrations: "))

manage_registrations(capacity, registrations)
```
