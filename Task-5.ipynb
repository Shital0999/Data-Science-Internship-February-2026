```python
# Problem 1: Social Media – Post Engagement Analyzer

def post_engagement_analyzer(likes):
    total_likes = 0

    # loop for calculation
    for like in likes:
        total_likes += like

    # condition check
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    return total_likes, status


# user input
n = int(input("Enter number of posts: "))
likes_list = []

for i in range(n):
    like = int(input(f"Enter likes for post {i+1}: "))
    likes_list.append(like)

# function call
total, status = post_engagement_analyzer(likes_list)

# output
print("Total Likes:", total)
print("Post Status:", status)
```

```python
def rainfall_checker():
    n = int(input())
    total = 0

    for i in range(n):
        value = float(input())
        total += value

    required = float(input())

    average = total / n

    if average >= required:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", round(average))
    print("Rainfall Status:", status)


rainfall_checker()
```

```python
# Problem 4: Social Media – Duplicate Account Detection

def duplicate_account_detection():
    n = int(input())
    usernames = []

    for i in range(n):
        name = input()
        usernames.append(name)

    if len(usernames) != len(set(usernames)):
        result = "Yes"

    else:
        result = "No"

    print("Duplicate Accounts Found:" , result)

duplicate_account_detection()
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter

def premium_crop_filter():
    n = int(input())
    prices = []

    for i in range(n):
        price = int(input())
        prices.append(price)

    premium = []

    for p in prices:
        if p > 2000:
            premium.append(p)
    print("Premium Crops:", premium)

premium_crop_filter()
        
```

```python
# Problem 7: System Monitoring – Application Health Checker

def application_health_checker():
    errors = int(input())

    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Tssues"

    print("Error Count:", errors)
    print("System Status:", status)

application_health_checker()
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker

def transaction_limit_checker():
    amount = int(input())
    limit = 50000

    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


transaction_limit_checker()
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility System

def attendance_eligibility():
    total_classes = int(input())
    attended_classes = int(input())

    percentage = (attended_classes / total_classes) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", round(percentage, 1))
    print("Exam Eligibility:", status)


attendance_eligibility()
```

```python
# Problem 10: Smart Electricity Bill Analyzer

def electricity_bill_analyzer():
    units = int(input())

    # bill calculation
    if units <= 100:
        bill = units * 3

    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5

    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    print("Total Bill:", bill)
    print("Usage Status:", status)


electricity_bill_analyzer()
```
