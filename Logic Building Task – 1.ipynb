```python
# Pass / Fail Analyzer
marks = [45,78,90,33,60]

pass_count = 0
fail_count = 0

# Loop through each student's marks
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

```python
# Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []

for name in names:
    clean_name = name.strip().lower()
    cleaned_names.append(clean_name)
    # strip() removes extra spaces
    # lower() converts text to lowercase

print("Cleaned Names:", cleaned_names)
    
```

```python
# Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

# Loop through each message
for message in messages:
    length = len(message)
    print("Message:", message)
    print("Length:", length)

    if length >10:
        print("Long Message Detected")
```

```python
# Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = logs.count("ERROR")

# Print total number of errors
print("Total ERROR entries:", error_count)
```
